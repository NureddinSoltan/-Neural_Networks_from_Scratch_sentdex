## What we're trying to do here?

we're trying to do inputs * weights + bias

### How can we multiply that?
we can use the do `products`

**`dot product:`** is you multiply elements with the values of these vectors and you add that all together.



```python
import numpy as np

inputs = [1, 2, 3, 2.5]
weights = [0.2, 0.8, -0.5, 1.0]
bias = 2

output = np.dot(weights, inputs) + bias
print(output)
```
