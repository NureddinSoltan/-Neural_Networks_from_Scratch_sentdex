# 1. Step function
the idea of **`step function`** is very simple that **if input of this function is greater than 0, then the output will be `one` otherwise it will be `zero`.**



## Taking the step function and use it as an activation function: the that's work in neural network:
- each neuron on inputs layer and outputs layer will have this activation function function.
- it will play after you put **inputs** times the **weights** + the **bias** ->(input* weight + bias) that value of what you being fit through the activation function.
- The output will be always just **zero** or **one**.
- The output of the activation function will be the input of the next neurons

# 3. ReLU (Rectified Linear Unit) activation function

the ReLU is pretty simple: **if x > 0 the out put will be whenever x is. and if x <= 0 the out put will be 0**

- it also comes after doing the **inputs** times the **weights** + the **bias** ->(input* weight + bias) like the sigmoid function, but it can flip the sign in terms of weights and biases.
- This is the most popular activation function to use for hidden layers and neural networks.

## 3.1 why we use something like ReLU over sigmoid?
- sigmoid has an issue for the gradient problem.
- it's fast and a very simple calculation. 
  - sigmoid 0 and 1.
  - ReLu 0 or x which is fast.
- it's just simply work.

# Code part

```python
import numpy as np

np.random.seed(0)

X = [[1, 2, 3, 2.5],
	[2.0, 5.0, -1.0, 2.0],
	[-1.5, 2.7, 3.3, -0.8]]

inputs = [0, 2, -1, 3.3, -2.7, 1.1, 2.2, -100]
output = []

# This is Rectified linear activation function
for i in inputs:
	if i > 0:
		output.append(i)
	elif i <= 0:
		output.append(0)

	# you can use that also it's the same:
	# output.append(max(0,i))

print(output)
```

## 4.1 Now let's right the ReLU activation function:

```python
import nnfs
from nnfs.datasets import spiral_data

nnfs.init()

X = [[1, 2, 3, 2.5],
	[2.0, 5.0, -1.0, 2.0],
	[-1.5, 2.7, 3.3, -0.8]]

X, y = spiral_data(100,3)

class Layer_Dense:
	def __init__(self, n_inputs, n_neurons):
		self.weights = 0.10 * np.random.randn(n_inputs, n_neurons) # randn gaussian distribution bounded around zero.
		self.biases = np.zeros((1, n_neurons)) # np.zeros is generating a zeros
	def forward(self, inputs):
		self.output = np.dot(inputs, self.weights) + self.biases

# Add Activation function
class Activation_ReLU:
	def forward(self, inputs):
		self.output = np.maximum(0, inputs)


layer1 = Layer_Dense(2,5)
activation1 = Activation_ReLU()
layer1.forward(X)

# print(layer1.output)
activation1.forward(layer1.output)
print(activation1.output)
```

## 4.2 Explaining the nnfs.datasets (spiral_data)

```python
def create_data(points, classes):
    X = np.zeros((points*classes, 2))
    y = np.zeros(points*classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(points*class_number, points*(class_number+1))
        r = np.linspace(0.0, 1, points)  # radius
        t = np.linspace(class_number*4, (class_number+1)*4, points) + np.random.randn(points)*0.2
        X[ix] = np.c_[r*np.sin(t*2.5), r*np.cos(t*2.5)]
        y[ix] = class_number
        print(f"Classes {classes}")
        print(f"Points {points}")
        print(f"x {X}")
        print(f"y {y}")
    return X, y


import matplotlib.pyplot as plt

print("here")
X, y = create_data(100,3)


plt.scatter(X[:, 0], X[:, 1], c=y, cmap="brg")
plt.show()

plt.scatter(X[:,0], X[:,1])
plt.show
```
